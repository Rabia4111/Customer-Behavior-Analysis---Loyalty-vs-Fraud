```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
```

```python
df = pd.read_csv('Dataset/synthetic_ecommerce_churn_dataset.csv')
df.head()
```

```python
df.info()
df.describe()
df.isna().sum()
```

```python
df['customer_since'] = pd.to_datetime(df['customer_since'])
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df['avg_order_value'] = df['avg_order_value'].fillna(df['avg_order_value'].median())
df['email_open_rate'] = df['email_open_rate'].fillna(df['email_open_rate'].median())
```

```python
from sklearn.preprocessing import LabelEncoder
import datetime

df['customer_since'] = pd.to_datetime(df['customer_since'])
today = pd.to_datetime('2025-07-14')
df['tenure_days'] = (today - df['customer_since']).dt.days

le_gender = LabelEncoder()
le_country = LabelEncoder()
le_category = LabelEncoder()

df['gender_encoded'] = le_gender.fit_transform(df['gender'])
df['country_encoded'] = le_country.fit_transform(df['country'])
df['category_encoded'] = le_category.fit_transform(df['preferred_category'])
```

```python
df_clean = df.drop(['customer_id', 'customer_since', 'gender', 'country', 'preferred_category'], axis=1)
df_clean.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='is_fraudulent', data=df)
plt.title("Fraudulent vs Legitimate Customers")
plt.show()
```

```python
sns.histplot(df['churn_risk'], bins=30, kde=True)
plt.title("Churn Risk Distribution")
plt.show()
```

```python
sns.boxplot(x='is_fraudulent', y='tenure_days', data=df)
plt.title("Customer Tenure vs Fraudulent Status")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='preferred_category', hue='is_fraudulent', data=df)
plt.title("Preferred Product Categories vs Fraudulent Status")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_clean.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
df_clean.to_csv("cleaned_customer_data.csv", index=False)
```
